```python
#Product Performance Analysis
```

```python
# Finding the best-selling products by summing the sales numbers of products
best_selling_products = data_set.groupby('Product type')['Number of products sold'].sum().reset_index()
best_selling_products = best_selling_products.sort_values(by='Number of products sold', ascending=False)
best_selling_products.head()
```

```python

# Sort the best-selling products dataframe by sales numbers in descending order
best_selling_products = best_selling_products.sort_values(by='Number of products sold', ascending=False)

# Create a histogram to visualize the best-selling products
plt.figure(figsize=(12, 6))
plt.bar(best_selling_products['Product type'][:10], best_selling_products['Number of products sold'][:10])
plt.xlabel('Product Type')
plt.ylabel('Total Sales')
plt.title('Top 10 Best-Selling Products')
plt.xticks(rotation=45)
plt.show()
```

```python
#Calculation of the Most Profitable Products:
```

```python
# Calculating the profitability of products (Number of products sold x Price)
data_set['Profit'] = data_set['Number of products sold'] * data_set['Price']
most_profitable_products = data_set.groupby('Product type')['Profit'].sum().reset_index()
most_profitable_products = most_profitable_products.sort_values(by='Profit', ascending=False)
most_profitable_products.head()
```

```python

top_10_most_profitable = most_profitable_products.head(10)

plt.figure(figsize=(8, 8))
plt.pie(top_10_most_profitable['Profit'], labels=top_10_most_profitable['Product type'], autopct='%1.1f%%', startangle=140)
plt.title('Top 10 Most Profitable Products')
plt.axis('equal')  
plt.show()
```

```python
#Visualisation of the Relationship between Inventory Levels and Sales:
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=data_set, x='Stock levels', y='Number of products sold', hue='Product type')
plt.title('Relationship Between Stock Levels and Sales')
plt.xlabel('Stock Levels')
plt.ylabel('Number of Products Sold')
plt.show()
```

```python
# Calculating the correlation between stock levels and sales
correlation = data_set['Stock levels'].corr(data_set['Number of products sold'])

# Printing the correlation coefficient
print(f'Correlation Coefficient: {correlation:.2f}')
```

```python
#Correlation coefficient (0.02) indicates a very weak relationship between inventory levels and sales.
#This suggests that inventory levels do not directly or significantly affect sales.
#Hence, it means that there is no strong relationship or dependency between inventory levels and sales.
```

```python
from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()
data_set['Inspection results_encoded'] = label_encoder.fit_transform(data_set['Inspection results'])
```

```python
data_set['Inspection results_encoded'].head()
```

```python
from scipy.stats import pearsonr
```

```python
# Evaluate the relationship between production volume and audit results
production_volumes = data_set['Production volumes']
inspection_results_encoded = data_set['Inspection results_encoded']

correlation, _ = pearsonr(production_volumes, inspection_results_encoded)

print(f"Pearson Correlation Coefficient: {correlation:.2f}")
```

```python
p_value = pearsonr(production_volumes, inspection_results_encoded)
print(f"Pearson Correlation Coefficient: {correlation:.2f}")
print(f"P-value: {p_value}")
```

```python
#The correlation between production volume and audit results 
#shows a very weak positive correlation
#and this correlation is not statistically significant.
```
